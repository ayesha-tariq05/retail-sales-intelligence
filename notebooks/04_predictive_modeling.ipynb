```python
import pandas as pd
from prophet import Prophet

# Load your cleaned data
df = pd.read_csv(r"d:\My_projects\Retail-sales-intelligence\data\processed\cleaned_sales_data.csv")

# Convert order_date and set it as index
df['order_date'] = pd.to_datetime(df['order_date'])
df_monthly = df.resample('ME', on='order_date')['sales'].sum().reset_index()

# Prophet requires specific column names: 'ds' for date, 'y' for the value
df_monthly.columns = ['ds', 'y']

print(df_monthly.head())
```

```python
# Train the Prophet model
model = Prophet()
model.fit(df_monthly)

# Create a dataframe for future dates (next 6 months)
future = model.make_future_dataframe(periods=6, freq='ME')
forecast = model.predict(future)

# Select only the relevant columns and save
forecast_data = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
forecast_data.to_csv(r"d:\My_projects\Retail-sales-intelligence\data\processed\forecasted_sales.csv", index=False)

print("Forecast generated and saved!")
```
