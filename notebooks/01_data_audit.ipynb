Data Cleaning

```python
# Use 'mixed' format to handle inconsistent date strings
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed')
df['ship_date'] = pd.to_datetime(df['ship_date'], format='mixed')

# Verify the changes
df.info()
```

```python
# Group by category and sum the sales
category_sales = df.groupby('category')['sales'].sum().sort_values(ascending=False)

# Display the results
print("Total Sales by Category:")
print(category_sales)
```

```python
import matplotlib.pyplot as plt

# Create the plot
category_sales.plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title('Total Sales by Category')
plt.ylabel('Total Sales ($)')
plt.xlabel('Product Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by category and sum both Sales and Profit
category_performance = df.groupby('category')[['sales', 'profit']].sum().sort_values(by='sales', ascending=False)

print("Performance by Category:")
print(category_performance)
```

```python
# Calculate Average Shipping Cost by Category
shipping_analysis = df.groupby('category')['shipping_cost'].mean().sort_values(ascending=False)

print("Average Shipping Cost per Category:")
print(shipping_analysis)
```

```python
# Calculate Average Discount by Category
discount_analysis = df.groupby('category')['discount'].mean().sort_values(ascending=False)

print("Average Discount per Category:")
print(discount_analysis)
```
