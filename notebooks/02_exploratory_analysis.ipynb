```python
import pandas as pd
import os

# Define the path
file_path = r"d:\My_projects\Retail-sales-intelligence\data\raw\SuperStoreOrders.csv"

# Load and clean in one go
df = pd.read_csv(file_path, engine='python', sep=None)

# Run the cleaning steps we established earlier
df['sales'] = df['sales'].replace({',': ''}, regex=True).astype(float)
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed')
df['ship_date'] = pd.to_datetime(df['ship_date'], format='mixed')

print("Data re-loaded and cleaned successfully!")
```

```python
# Group by region and sum profit
regional_profit = df.groupby('region')['profit'].sum().sort_values(ascending=False)

print("Total Profit by Region:")
print(regional_profit)
```

```python
# Group by region and sum profit
regional_profit = df.groupby('region')['profit'].sum().sort_values(ascending=False)

print("Total Profit by Region:")
print(regional_profit)
```
